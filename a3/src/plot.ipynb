```python
from triest import *
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
Ms = [500, 1000, 2000, 5000, 10000]
estimations_base = [[] for _ in range(len(Ms))]
DATASET_PATH = "../dataset/facebook_combined.txt"
for i, M in enumerate(Ms):
    print(f"# M={M} #")
    for _ in tqdm(range(10)):
        triest_alg = TriestBase(M, DATASET_PATH)
        tau = triest_alg.process()
        estimations_base[i].append(tau)
```

```python
x_positions = range(1, len(Ms) + 1)

plt.figure(figsize=(12, 8))
for i, experiment_results in enumerate(estimations_base):
    plt.scatter([x_positions[i]]*len(experiment_results), experiment_results)

plt.plot([1, 5], [1612010, 1612010], color='black')
plt.xticks(x_positions, Ms)
plt.grid()
plt.title("Triest Base: Estimation of Number of Triangles VS Size of Reservoir")
plt.xlabel("Size of Reservoir")
plt.ylabel("Estimation of Number of Triangles")
plt.savefig("../fig/result_base.jpg", dpi=300)
```

```python
x_positions = range(1, len(Ms))

plt.figure(figsize=(12, 8))
for i, experiment_results in enumerate(estimations_base[1:]):
    plt.scatter([x_positions[i]]*len(experiment_results), experiment_results)

plt.plot([1, 4], [1612010, 1612010], color='black')
plt.xticks(x_positions, Ms[1:])
plt.grid()
plt.title("Triest Base: Estimation of Number of Triangles VS Size of Reservoir")
plt.xlabel("Size of Reservoir")
plt.ylabel("Estimation of Number of Triangles")
plt.savefig("../fig/result_base_zoom.jpg", dpi=300)
```
